```python
import PIL
from PIL import Image
import numpy as np
from PIL import ImageFilter
import matplotlib.pyplot as plt
import cv2
from matplotlib.pyplot import imshow
```

```python
def convert_to_grayscale(image):
    #Converting the Image into the Grayscale image
    rgb = np.asarray(image)
    #Seperating the R,G,B arrays into separates r,g,b numpy array
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]

    #finally adding all r,g,b to gray plus multiplying with some parameters
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

# Task 1 

```python
image = cv2.imread('lena_color.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image.shape
```

## The original RGB Image

```python

imshow(image)
```

## Converting RBG into its inverse

```python
#converting rgb image into its inverse image
imshow(255  - image)
```

## Converting Grayscale image into its inverse

```python
#getting the gray scale image
gray_image = convert_to_grayscale(image)
imshow (255 - image)
```

## Converting the Binarized image into its inverse

```python
#Loadingt the image
im_binary = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
#Subtracting the values with the maximum intensities in this case 1
imshow(1 - im_binary[1], cmap = 'gray')
```

## Task 2

## Gradient along the x-axis

```python
#getting the number of rows and columns
rows = image.shape[0]
col = image.shape[1]


gradient_image_x = np.zeros((rows, col))
for i in range(rows):
    for j in range(col):
        #catering the index out of bound error
        if (i < 511):
            #abs if the value is negative
            gradient_image_x[i][j] = abs(gray_image[i + 1][j] - gray_image[i][j])


imshow(gradient_image_x, cmap = "gray")
            
        
```

## Gradient along the y-axis

```python
rows = image.shape[0]
col = image.shape[1]


gradient_image_y = np.zeros((rows, col))
for i in range(rows):
    for j in range(col):
        if (j < 511):
            gradient_image_y[i][j] = abs(gray_image[i][j+1] - gray_image[i][j])


imshow(gradient_image_y, cmap = "gray")
            
```

## Task 3

```python
# Reading the dollar image
dollar_image = cv2.imread('dollar.png')
dollar_image = cv2.cvtColor(dollar_image, cv2.COLOR_BGR2RGB)
#conversion of the orignal image to grayscale image
dollar_grayimage = convert_to_grayscale(dollar_image)

dollar_grayimage.shape
dollar_grayimage
```

## Original Image

```python
imshow(dollar_image)
```

## Generating the binary bit representation array of grayscale image 

```python
 
stringBinary = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))

for i in range(dollar_grayimage.shape[0]):
    for j in range(dollar_grayimage.shape[1]):
        #"{0:b}".format(x) => return the bits representation of int x
        stringBinary[i][j] = "{0:b}".format(int(dollar_grayimage[i][j]))
        
```

```python
#outputing the bits array
stringBinary
```

## Generating empty arrays of size image

```python
plane1 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane2 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane3 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane4 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane5 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane6 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane7 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
plane8 = np.zeros((dollar_image.shape[0], dollar_image.shape[1]))
```

## Putting the bits into respective planes

```python
for row in range(dollar_image.shape[0]):
    for col in range(dollar_image.shape[1]):
        #minimum length of the string value is 3 i.e 0.0 => len(0.0) = 3
        if (len(str(stringBinary[row][col]))  == 3):
            plane1[row][col] = str(stringBinary[row][col])[-3]
        
        if (len(str(stringBinary[row][col])) > 3 ):
            plane2[row][col] = str(stringBinary[row][col])[-4]
        
        if (len(str(stringBinary[row][col])) >4):
            plane3[row][col] = str(stringBinary[row][col])[-5]
        
        if (len(str(stringBinary[row][col])) >5):
            plane4[row][col] = str(stringBinary[row][col])[-6]
        
        if (len(str(stringBinary[row][col])) >6):
            plane5[row][col] = str(stringBinary[row][col])[-7]
        
        if (len(str(stringBinary[row][col])) >7):
            plane6[row][col] = str(stringBinary[row][col])[-8]
        
        if (len(str(stringBinary[row][col])) >8):
            plane7[row][col] = str(stringBinary[row][col])[-9]
        
        if (len(str(stringBinary[row][col])) >9):
            plane8[row][col] = str(stringBinary[row][col])[-10]
        
```
