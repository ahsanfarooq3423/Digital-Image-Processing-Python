```python
import PIL
from PIL import Image
import numpy as np
from PIL import ImageFilter
import matplotlib.pyplot as plt
import cv2
from matplotlib.pyplot import imshow
```

```python
img = cv2.imread('image.jpg')
img.shape
```

```python
def plot(img):
    arr=np.asarray(img)
    plt.hist(arr.ravel(),bins=256,range=(0,255),fc='k',ec='k')
    plt.show()
```

## The original low contrast image with its histogram

```python
imshow(img)
```

#### Original Image Histogram

```python
plot(img)
```

### Technique 1 -> CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
image = cv2.imread('image.jpg', 0)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(image)
res = np.hstack((image,cl1)) 
imshow(res, cmap = 'gray', aspect = 'auto')
```

```python
plot(cl1)
```

### Technique 2 -> Histogram Equalization 

```python
equ = cv2.equalizeHist(image)
res = np.hstack((image,equ)) 
imshow(res, cmap = 'gray', aspect = 'auto')
```

```python
plot(equ)
```
