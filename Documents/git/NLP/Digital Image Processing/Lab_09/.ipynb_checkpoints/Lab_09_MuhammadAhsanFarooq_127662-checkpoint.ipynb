# Task 1: Averaging Filter

```python
def averagingFilter(image, n):
    arr=numpy.asarray(image)
    mask=numpy.ones(shape=(n,n))
    inc=int(n/2)
    narr=numpy.pad(arr,pad_width=inc,mode='constant') #padding with zeros
    smooth=numpy.zeros(shape=(len(arr),len(arr[0]))).tolist()

    for x in range(0+inc,len(arr)+inc):
        for y in range(0+inc,len(arr[0])+inc):
            fill=narr[x-inc:x+inc+1,y-inc:y+inc+1]
            fil= numpy.multiply(fill,mask)
            sum=numpy.sum(fill)
            avg=sum/(n*n)
            smooth[x-inc][y-inc]=avg
    return smooth
```

```python
img=Image.open('smoothing.tif')
img = img.convert('RGBA')
img = img.convert('L')
plt.imshow(img)
plt.xlabel('Original Image')
imshow(img, cmap = 'gray')
```

```python
smooth_img_array_3 = averaging(img, 3)
smooth_img_3=Image.fromarray(numpy.array(smooth_img_array_3))
plt.imshow(smooth_img_3)
plt.xlabel('Filter of size 3 x 3')
plt.show()
```

# Task 2: Gaussian Smoothing

```python
img=Image.open('smoothing.tif')
img = img.convert('RGBA')
img = img.convert('L')
```

```python
def gaussian_smoothing(image, n):
    arr=numpy.asarray(img)
    mask=numpy.array([[1,1,2,2,2,1,1],[1,2,2,4,2,2,1],[2,2,4,8,4,2,2],[2,4,8,16,8,4,2],[2,2,4,8,4,2,2],[1,2,2,4,2,2,1],[1,1,2,2,2,1,1]])
    inc=int(n/2)
    narr=numpy.pad(arr,pad_width=inc,mode='constant') #padding with zeros
    smooth=numpy.zeros(shape=(len(arr),len(arr[0]))).tolist()

    for x in range(0+inc,len(arr)+inc):
        for y in range(0+inc,len(arr[0])+inc):
            fill=narr[x-inc:x+inc+1,y-inc:y+inc+1]
            fill=numpy.multiply(fill,mask)
            sum=numpy.sum(fill)
            avg=sum/(mask.sum())
            smooth[x-inc][y-inc]=avg
    return smooth
```

```python
output_img_arr = gaussian_smoothing(img, 7)
output_img=Image.fromarray(numpy.array(output_img_arr))
plt.imshow(output_img)
plt.show()
```

# Task 4: Removal of Salt and Pepper Noise

```python
def median(image,n):
    arr=numpy.asarray(image)
    inc=int(n/2)
    narr=numpy.pad(arr,pad_width=inc,mode='constant') #padding with zeros
    enhanced=numpy.zeros(shape=(len(arr),len(arr[0]))).tolist()

    for x in range(0+inc,len(arr)+inc):
        for y in range(0+inc,len(arr[0])+inc):
            window=narr[x-inc:x+inc+1,y-inc:y+inc+1]
            enhanced[x-inc][y-inc]=numpy.median(window)
    return enhanced
```

```python
image_2 =Image.open('saltandpaper.tif')
image_2 = image_2.convert('RGBA')
image_2 = image_2.convert('L')
plt.imshow(image_2)
plt.xlabel('Original Image')
plt.show()
```

```python
noise_removal_array = median(image_2, 7)
noise_removal_image = Image.fromarray(numpy.array(noise_removal_array))
plt.imshow(noise_removal_image)
plt.title('Noise Removal Using Median Filter')
plt.show()
```
