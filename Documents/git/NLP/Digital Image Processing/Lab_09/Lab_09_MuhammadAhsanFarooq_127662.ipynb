## Task 1

```python
def averagingFilter(image, n):
    arr=np.asarray(image)
    kernel=np.ones(shape=(n,n))
    inc=int(n/2)
    narr=np.pad(arr,pad_width=inc,mode='constant') #padding with zeros
    final=np.zeros(shape=(len(arr),len(arr[0]))).tolist()

    for x in range(0+inc,len(arr)+inc):
        for y in range(0+inc,len(arr[0])+inc):
            fill=narr[x-inc:x+inc+1,y-inc:y+inc+1]
            fil= np.multiply(fill,kernel)
            sum=np.sum(fill)
            avg=sum/(n*n)
            final[x-inc][y-inc]=avg
    return final
```

#### Original Image

```python
img=Image.open('smoothing.tif')
img = img.convert('RGBA')
img = img.convert('L')
imshow(img, cmap ='gray')
```

#### Averaging Filter with 3x3 filter

```python
img_3_3 = averagingFilter(img, 3)
imshow(img_3_3, cmap = 'gray')
```

#### Averaging Filter with 5x5 filter

```python
img_5_5 = averagingFilter(img, 5)
imshow(img_5_5, cmap = 'gray')
```

#### Averaging Filter with 15x15 filter

```python
img_15_15 = averagingFilter(img, 15)
imshow(img_15_15, cmap = 'gray')
```

#### Averaging Filter with 35x35 filter

```python
img_35_35 = averagingFilter(img, 35)
imshow(img_35_35, cmap = 'gray')
```

### What do you observe when increasing the size of the filter and why?

##### Ans: By increasing the size of the filter the images get more blurred because in a bigger size filter we are taking the average of more neighbourhood pixels thus removing more noise

```python
img=Image.open('smoothing.tif')
img = img.convert('RGBA')
img = img.convert('L')
```

```python
def gaussianSmoothing(image, n):
    arr=np.asarray(img)
    mask=np.array([[1,1,2,2,2,1,1],[1,2,2,4,2,2,1],[2,2,4,8,4,2,2],[2,4,8,16,8,4,2],[2,2,4,8,4,2,2],[1,2,2,4,2,2,1],[1,1,2,2,2,1,1]])
    inc=int(n/2)
    narr=np.pad(arr,pad_width=inc,mode='constant') 
    final=np.zeros(shape=(len(arr),len(arr[0]))).tolist()

    for x in range(0+inc,len(arr)+inc):
        for y in range(0+inc,len(arr[0])+inc):
            fill=narr[x-inc:x+inc+1,y-inc:y+inc+1]
            fill=np.multiply(fill,mask)
            sum=np.sum(fill)
            avg=sum/(mask.sum())
            final[x-inc][y-inc]=avg
    return final
```

```python
guass_img = gaussianSmoothing(img, 7)
imshow(guass_img, cmap = 'gray')
```
